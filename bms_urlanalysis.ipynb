```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, roc_curve, confusion_matrix
from sklearn.ensemble import VotingClassifier
from lightgbm import LGBMClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Load Dataset
data = pd.read_csv('C:/Users/rthar/Documents/projects/datasets/dataset.csv')

# Data Overview
print(data.info())
print(data.head())

# Check for Null Values and Unique Counts
print("Null values in data:\n", data.isnull().sum())
print("Unique elements in each column:\n", data.nunique())

# Drop irrelevant features
data_reduced = data.drop(['popUpWidnow'], axis=1)

# Features and Target Split
X = data_reduced.drop(['index', 'Result'], axis=1)
y = data_reduced['Result']

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Data Normalization/Standardization (Optional, depending on dataset)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Model Definitions
# LightGBM for gradient boosting
lgbm = LGBMClassifier(
    boosting_type='gbdt',
    objective='binary',
    learning_rate=0.1,
    n_estimators=200,
    random_state=42
)

# Logistic Regression
lr = LogisticRegression(max_iter=1000, random_state=42)

# Random Forest
rf = RandomForestClassifier(n_estimators=150, max_depth=10, random_state=42)

# Ensemble: Voting Classifier
voting_clf = VotingClassifier(
    estimators=[
        ('LightGBM', lgbm),
        ('LogisticRegression', lr),
        ('RandomForest', rf)
    ],
    voting='soft'  # Soft voting for probability-based predictions
)

# Cross-Validation Evaluation
skf = StratifiedKFold(n_splits=5)
lgbm_scores = []
rf_scores = []
lr_scores = []

for train_index, val_index in skf.split(X_train, y_train):
    X_train_fold, X_val_fold = X_train[train_index], X_train[val_index]
    y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]

    lgbm.fit(X_train_fold, y_train_fold)
    rf.fit(X_train_fold, y_train_fold)
    lr.fit(X_train_fold, y_train_fold)

    lgbm_scores.append(accuracy_score(y_val_fold, lgbm.predict(X_val_fold)))
    rf_scores.append(accuracy_score(y_val_fold, rf.predict(X_val_fold)))
    lr_scores.append(accuracy_score(y_val_fold, lr.predict(X_val_fold)))

print("LightGBM CV Accuracy:", np.mean(lgbm_scores))
print("Random Forest CV Accuracy:", np.mean(rf_scores))
print("Logistic Regression CV Accuracy:", np.mean(lr_scores))

# Final Model Training
voting_clf.fit(X_train, y_train)

# Predictions
y_pred = voting_clf.predict(X_test)
y_probs = voting_clf.predict_proba(X_test)[:, 1]

# Model Evaluation
print("Test Set Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ROC Curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
roc_auc = roc_auc_score(y_test, y_probs)

plt.figure(figsize=(10, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend(loc="lower right")
plt.show()

# Save the Final Model
import joblib
joblib.dump(voting_clf, 'voting_classifier_model.pkl')
```
